```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Age': [12, 18, 25, 33, 42, 48],
    'Attendance (%)': [85, 90, 95, 80, 75, 65]
}

df = pd.DataFrame(data)

bins = [10, 15, 30, 40, 50]
labels = ['10-15', '15-30', '30-40', '40-50']

df['Age_group'] = pd.cut(df['Age'], bins=bins, labels=labels)

grouped = df.groupby('Age_group')['Attendance (%)'].mean()
print(grouped)

```
